# Repository-oriented clustering

```python
bot_activities = (
    activities[activities['category'] == 'bot']
    .pivot_table(
        index=['repository', 'contributor'],
        columns='activity',
        aggfunc='size',
        fill_value=0
    )
    .reset_index()
    .drop_duplicates(
        subset=['repository'], 
        keep='first', 
        inplace=False, 
        ignore_index=True
    )
)
bot_activities
```

```python
bot_activities['contributor'].value_counts()
```

# Contributor-oriented clustering

## Taking into account repositories (separated)

```python
bot_activities = (
    activities[activities['category'] == 'bot']
    .pivot_table(
        index=['contributor', 'repository'], 
        columns='activity', 
        aggfunc='size', 
        fill_value=0
    )
    .reset_index()
)
bot_activities
```

```python
human_activities = (
    activities[activities['category'] == 'human']
    .pivot_table(
        index=['contributor', 'repository'], 
        columns='activity', 
        aggfunc='size', 
        fill_value=0
    )
    .reset_index()
)
human_activities
```

## Without taking into account repositories

```python
bot_activities = (
    activities[activities['category'] == 'bot']
    .pivot_table(
        index=['contributor'], 
        columns='activity', 
        aggfunc='size', 
        fill_value=0
    )
    .reset_index()
    .set_index('contributor')
)
bot_activities
```

```python
sns.clustermap(bot_activities.sample(20), method='centroid')
```
