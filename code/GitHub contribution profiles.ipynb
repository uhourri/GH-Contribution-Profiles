# Data importing

```python
import pandas as pd
import numpy as np
```

```python
activities = pd.read_csv('../data/activities.csv')
activities
```

# Contributor clustering

## Taking into account repositories (separated)

```python
bot_activities = (
    activities[activities['category'] == 'bot']
    .pivot_table(
        index=['contributor', 'repository'], 
        columns='activity', 
        aggfunc='size', 
        fill_value=0
    )
    .reset_index()
)
bot_activities
```

```python
human_activities = (
    activities[activities['category'] == 'human']
    .pivot_table(
        index=['contributor', 'repository'], 
        columns='activity', 
        aggfunc='size', 
        fill_value=0
    )
    .reset_index()
)
human_activities
```

## Without taking into account repositories

```python
bot_activities = (
    activities[activities['category'] == 'bot']
    .pivot_table(
        index=['contributor'], 
        columns='activity', 
        aggfunc='size', 
        fill_value=0
    )
    .reset_index()
)
bot_activities
```
